# Frequency Modulation in the Time Domain

Import necessary libraries

```
import numpy as np
import matplotlib.pyplot as plt
import math
from scipy import signal
```

Set the time variable, we will generate the signal from 0 to 30m sec

```
Tstart=0
Tstop=3e-2
Tstep=1e-5

t=np.arange(Tstart,Tstop,Tstep)
```

Define the message signal $m(t)=A_mcos(2 \pi f_mt)$

```
Am=5    # amplitude of message signal
fm=100  # frequency of message signal
mt=Am*np.cos(2*math.pi*fm*t)  # message signal
```

Plot the message signal $m(t)$

```
plt.plot(t,mt)
```

Now, produce the carrier signal $c(t)=A_ccos(2 \pi f_ct)$

```
Ac=5      # amplitude of carrier signal
fc=1000   # frequency of carrier signal
ct=Ac*np.cos(2*math.pi*fc*t)
```

Plot the carrier signal $c(t)$

```
plt.plot(t,ct,'k')
```

Let us now produce the FM signal $s(t)=A_c \cos\left( 2 \pi f_c t+ k_f \int_0^t m(\tau) d\tau \right) $. 

When the modulating signal $m(t)$ is sinusoidal, the FM signal can be expressed as: $s(t)=A_c \cos\left( 2 \pi f_c t+\beta \sin(2\pi f_m t) \right)$, where $\beta = \frac{k_f A_m}{f_m}$.

```
Kf=100      # frequency sensitivity (Hz/volt)
B= Kf*Am/fm # beta, frequency modulation index
st=Ac*np.cos(2*math.pi*fc*t + B*np.sin(2*math.pi*fm*t))   # modulated signal
```

Let us plot the modulated signal s(t)

```
plt.plot(t,st,'r')
```

As can be observed from the above figure, the modulated signal has constant amplitude and varying instantaneous frequency. To determine the relationship between the original message and the modulated signal, let us plot the message signal $m(t)=\cos(2\pi f_m t)$ with the modulated signal $s(t)=A_c \cos\left( 2 \pi f_c t+\beta \sin(2\pi f_m t) \right)$

```
plt.plot(t,mt,'b--')
plt.plot(t,st,'r')
```

It can be observed from the curves in the above figure, that the instantaneous frequency of the modulated signal (red curve) increases with the increase of the amplitude of the original message (blue curve). To investigate the FM modulation parameters further, let us put all the code together

```
Am=5     # amplitude of the message signal 
fm=100     # frequency of the message signal 
Ac=5       # amplitude of the carrier signal 
fc=1000    # frequency of the carrier signal 
Kf=100    # frequency-snsitivity factor

Tstart=0   # setting the run-time (t-axis)
Tstop=3e-2
Tstep=1e-5
t=np.arange(Tstart,Tstop,Tstep)

mt=Am*np.cos(2*math.pi*fm*t)    # message signal 
ct=Ac*np.cos(2*math.pi*fc*t)    # carrier signal

B= Kf*Am/fm     # FM index
st=Ac*np.cos(2*math.pi*fc*t + B*np.sin(2*math.pi*fm*t)) # modulated signal (FM signal)

plt.plot(t,mt,'b--')  # plot the message signal (modulating signal)
plt.plot(t,st,'r')    # plot the FM signal (modulated signal)
```

Now, let us vary the value of $A_m$ and see how it affects the FM modulated signal.

```
# Changing Am
Am1=9     # amplitude of the first message signal 
Am2=5     # amplitude of the second message signal 
Am3=1     # amplitude of the third message signal 

fm=100     # frequency of the message signal 
Ac=5       # amplitude of the carrier signal 
fc=1000    # frequency of the carrier signal 
Kf=100    # frequency-snsitivity factor

Tstart=0   # setting the run-time (t-axis)
Tstop=3e-2
Tstep=1e-5
t=np.arange(Tstart,Tstop,Tstep)


f = plt.figure(figsize=(20,4)) # to plot figures
ax1 = f.add_subplot(131)
ax2 = f.add_subplot(132)
ax3 = f.add_subplot(133)

ct=Ac*np.cos(2*math.pi*fc*t)    # carrier signal

mt1=Am1*np.cos(2*math.pi*fm*t)    # message signal 
B1= Kf*Am1/fm     # FM modulation index
st1=Ac*np.cos(2*math.pi*fc*t + B1*np.sin(2*math.pi*fm*t)) # modulated signal (FM signal)

mt2=Am2*np.cos(2*math.pi*fm*t)    # message signal 
B2= Kf*Am2/fm     # FM modulation index
st2=Ac*np.cos(2*math.pi*fc*t + B2*np.sin(2*math.pi*fm*t)) # modulated signal (FM signal)


mt3=Am3*np.cos(2*math.pi*fm*t)    # message signal 
B3= Kf*Am3/fm     # FM modulation index
st3=Ac*np.cos(2*math.pi*fc*t + B3*np.sin(2*math.pi*fm*t)) # modulated signal (FM signal)

ax1.plot(t,mt1,'b--')  # plot the message signal (modulating signal)
ax1.plot(t,st1,'r')    # plot the FM signal (modulated signal)
ax1.set_title('$A_m ={}$'.format(Am1))

ax2.plot(t,mt2,'b--')  # plot the message signal (modulating signal)
ax2.plot(t,st2,'r')    # plot the FM signal (modulated signal)
ax2.set_title('$A_m ={}$'.format(Am2))

ax3.plot(t,mt3,'b--')  # plot the message signal (modulating signal)
ax3.plot(t,st3,'r')    # plot the FM signal (modulated signal)
ax3.set_title('$A_m ={}$'.format(Am3))
```

# FM in the Frequency Domain

Let us define $m(t), c(t)$ and $s(t)$ as we did in the previous section

```
Am=1     # amplitude of the message signal 
fm=100     # frequency of the message signal 
Ac=5       # amplitude of the carrier signal 
fc=1000    # frequency of the carrier signal 
Kf=100    # frequency-snsitivity factor

Tstart=0    # to set the frequency axis
Tstop=10000/fc
Tstep=0.3/fc
t=np.arange(Tstart,Tstop,Tstep)
N=len(t) 

mt=Am*np.cos(2*math.pi*fm*t)    # message signal 
ct=Ac*np.cos(2*math.pi*fc*t)    # carrier signal

B= Kf*Am/fm     # FM modulation index
st=Ac*np.cos(2*math.pi*fc*t + B*np.sin(2*math.pi*fm*t)) # modulated signal (FM signal)
print('FM modulation index = {}'.format(B))
```

Let us compute and plot the amplitude spectrum of m(t). We will use
the np.fft.fft() method.

```
Mf = np.abs(np.fft.fft(mt))
freqs = np.fft.fftfreq(mt.size, Tstep)
idx = np.argsort(freqs)
freqs=freqs[idx]
Mf=Mf[idx]/(N)
plt.plot(freqs, Mf,'b')
plt.xlabel('freq (Hz)')
plt.ylabel('Amplitude Spectrum (v/Hz)')
```

Let us also compute and plot the amplitude spectrum of c(t).

```
Cf = np.abs(np.fft.fft(ct))         # fft of carrier signal
freqs = np.fft.fftfreq(ct.size, Tstep)
idx = np.argsort(freqs)
freqs=freqs[idx]
Cf=Cf[idx]/N
plt.plot(freqs, Cf,'k')
plt.xlabel('freq (Hz)')
plt.ylabel('Amplitude Spectrum (v/Hz)')
```

Now, let us compute and plot the amplitude spectrum of the FM modulated signal s(t).

```
Sf = np.abs(np.fft.fft(st))         # fft of modulated signal
freqs = np.fft.fftfreq(st.size, Tstep)
idx = np.argsort(freqs)
freqs=freqs[idx]
Sf=Sf[idx]/N
plt.plot(freqs, Sf,'r')
plt.xlabel('freq (Hz)')
plt.ylabel('Amplitude Spectrum (v/Hz)')
```

Let us zoom in at $f_c$

```
Sf = np.abs(np.fft.fft(st))         # fft of modulated signal
freqs = np.fft.fftfreq(st.size, Tstep)
idx = np.argsort(freqs)
freqs=freqs[idx]
Sf=Sf[idx]/N
plt.plot(freqs, Sf,'r')
plt.xlabel('freq (Hz)')
plt.ylabel('Amplitude Spectrum (v/Hz)')
```

# FM Demodulation

Let us use the discriminator to demodulate the FM signal. The discriminator used here is a differentator followed by an envelope detector.

The output of the differentiator is $\frac{ds}{dt}=-2 \pi A_c \left[  f_c +k_f  m(t) \right] \sin \left( 2\pi f_c t +k_f \int_0^t m(\tau)d\tau \right)$

$\frac{ds}{dt}=-2 \pi A_c \left[  f_c +k_f  Am \cos(2\pi f_m t) \right] \sin \left( 2\pi f_c t +\beta \sin(2 \pi f_m t   )\right)$



```
# generate the FM signal
Am=5    # amplitude of the message signal 
fm=100     # frequency of the message signal 
Ac=5       # amplitude of the carrier signal 
fc=1000    # frequency of the carrier signal 
Kf=100    # frequency-snsitivity factor

Tstart=0   # setting the run-time (t-axis)
Tstop=2e-2
Tstep=1e-4
t=np.arange(Tstart,Tstop,Tstep)

mt=Am*np.cos(2*math.pi*fm*t)    # message signal 
ct=Ac*np.cos(2*math.pi*fc*t)    # carrier signal
B= Kf*Am/fm                       # FM modulation index
st=Ac*np.cos(2*math.pi*fc*t + B*np.sin(2*math.pi*fm*t)) # modulated signal (FM signal)
ds= -2*math.pi*Ac*(fc + Kf* mt)*np.sin(2*math.pi*fc*t +B*np.sin(2*math.pi*fm*t))
plt.plot(t,ds,'r')
```

The signal at the output of the discriminator is an amplitude modulated (AM) signal. Consequently, if $f_c$ is large enough such that the carrier is not overmodulated, then we can recover the message signal with an envelope detector in a manner similar to that in AM signals. 

Let us now restore the message signal by removing the DC component of the signal at the output of the discriminator

```
At= np.abs(-2*math.pi*Ac*(fc + Kf* mt) )    # signal after envelope detector (before removing DE component)
dc= At - np.mean(At)                # the DC component
plt.plot(t,At,'b')
plt.plot(t,dc,'k')
plt.plot(t,ds,'r')
```

# FM of Square Wave

Let us now plot the FM signal when the modulating signal $m(t)$ is a square wave.

```
# generate the FM signal
Am=4
fm=100     # frequency of the message signal 
Ac=4       # amplitude of the carrier signal 
fc=1000    # frequency of the carrier signal 
Kf=100     # frequency-snsitivity factor

Tstart=0   # setting the run-time (t-axis)
Tstop=2e-2
Tstep=1e-5
t=np.arange(Tstart,Tstop,Tstep)

mt= Am* (signal.square(2 * np.pi * fm * t))
ct=Ac*np.cos(2*math.pi*fc*t)
st=Ac*np.cos(2*math.pi*fc*t + 2*Kf*math.pi*(mt)*t)

plt.plot(t,mt,'b--' )
plt.plot(t,st,'r')
plt.xlabel('t (sec)')
plt.ylabel('$s(t), m(t)$ (v)')
plt.title('FM of a square wave')

```

# Phase Modulation

The phase modulated signal is given by: $s(t)=A_c \cos\left( 2 \pi f_c t+k_p m(t) \right)$

Let us now look at the PM signal for the case when the message signal is sinusoidal. $s(t)=A_c \cos\left( 2 \pi f_c t+k_p A_m \cos(2\pi f_m t) \right)$. The result is compared to the FM signal.

```
Am=5     # amplitude of the message signal 
fm=100     # frequency of the message signal 
Ac=5       # amplitude of the carrier signal 
fc=1000    # frequency of the carrier signal 
Kp=1    # frequency-snsitivity factor

Tstart=0   # setting the run-time (t-axis)
Tstop=3e-2
Tstep=1e-5
t=np.arange(Tstart,Tstop,Tstep)

mt=Am*np.cos(2*math.pi*fm*t)    # message signal 
ct=Ac*np.cos(2*math.pi*fc*t)    # carrier signal


f = plt.figure(figsize=(15,4))
ax1 = f.add_subplot(121)
ax2 = f.add_subplot(122)

B= Kf*Am/fm     # FM modulation index
FM_st= Ac*np.cos(2*math.pi*fc*t + B*np.sin(2*math.pi*fm*t)) # modulated signal (FM signal)
PM_st= Ac*np.cos(2*math.pi*fc*t + Kp*mt) # modulated signal (PM signal)

ax1.plot(t,mt,'b--')  # plot the message signal (modulating signal)
ax1.plot(t,PM_st,'r') # plot the FM signal (modulated signal)
ax1.set_xlabel('t (sec)')
ax1.set_ylabel('$m(t)$ (v)')
ax1.set_title('PM signal')

ax2.plot(t,mt,'b--')  # plot the message signal (modulating signal)
ax2.plot(t,FM_st,'r') # plot the FM signal (modulated signal)
ax2.set_xlabel('t (sec)')
ax2.set_ylabel('$m(t)$ (v)')
ax2.set_title('FM signal')
```
